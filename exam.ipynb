```python
# Restaurant Menu System  Use a dictionary for items and prices. Ask user to order items, then display total cost.

items = {
    "Pizza" : 120,
    "Pasta" : 100,
    "Burger" : 60,
    "Maggi" : 100,
    "ColdCoffee" : 150
}

print("1. for Pizza")
print("2. for Pasta")
print("3. for Burger")
print("4. for Maggi")
print("5. for ColdCoffee")
print("6. End")

total = 0

while True:
    c = int(input("Enter the number of item you want to order :-"))
    
    if c == 1:
        p = items["Pizza"]
    elif c == 2:
        p = items["Pasta"]
    elif c == 3:
        p = items["Burger"]
    elif c == 4:
        p = items["Maggi"]
    elif c == 5:
        p = items["ColdCoffee"]
    elif c == 6:
        break

    total += p

print("Total Bill :-",total)
```

```python
# Password Strength Checker  Input a password and check if it contains at least one uppercase, one lowercase, one digit, and one special character.
password = input("Enter your password")

length = len(password)
has_digit = 0
has_upper = 0
has_special = 0

for ch in password:
    if ch >= '0' and ch <= '9':
        has_digit = 1
    elif ch >= 'A' and ch <= 'Z':
        has_upper = 1
    elif not (ch >= 'a' and ch <= 'z'):
        has_special = 1

if length < 6:
    print("Weak Password")
elif has_digit and has_upper and has_special:
    print("Strong Password")
else:
    print("Medium Password")
```

```python
# Online Store Inventory  Maintain a dictionary of items with quantity and price. Allow user to buy items, update stock, and print final bill.
# print("\n1. for buy items and 2. for final bill with order details and stock update. and exit")

items = {
    "T-Shirt" : {"quantity" : 10, "price" : 499},
    "Shirt" : {"quantity" : 20, "price" : 999},
    "Pent" : {"quantity" : 14, "price" : 799},
    "Shoes" : {"quantity" : 30, "price" : 1199},
    "Mask" : {"quantity" : 13, "price" : 199},
}

bought = {}

total = 0

while True:
    print("\n1. Buy Items")
    print("2. Final Bill")
    c = int(input("Enter your choice 1 or 2:-"))

    if c == 1:
        item = input("Enter the name of item :-")
        q = int(input("Enter the quantity :-"))

        if q > items[item]["quantity"]:
            print("\nStock is not available.")
        else:
            items[item]["quantity"] -= q
            total += q * items[item]["price"]
            if item not in bought:
                bought[item] = {"quantity" : q, "price" : items[item]["price"]}
            else:
                bought[item] = {"quantity" : bought[item]["quantity"] + q, "price" : items[item]["price"]}
            
            print("Purchase Done!!")

    elif c == 2:

        print("\nOrder Details :-")

        print(f" | {"Name" :<10} | {"Quantity":<10} | {"Price" :<10} | {"Total" :<10}")
        for i in bought:
            print(f" | {i :<10} | {bought[i]["quantity"] :<10} | {bought[i]["price"] :<10} | {bought[i]["quantity"] * bought[i]["price"] :<10}")

        print("\nFinal Bill :-",total)

        print("\nInventory Details :-")
        print(f" | {"Name" :<10} | {"Quantity":<10} | {"Price" :<10} |")
        for i in items:
            print(f" | {i :<10} | {items[i]["quantity"] :<10} | {items[i]["price"] :<10} | ")

        break

    else:
        print("\nError :- Enter valid number.")
```

```python
# Movie Ratings Aggregator  Use a dictionary of {movie: [ratings]}. Write a function to calculate the average rating for each movie.

r = {
    "Inception" : [5.0, 2.0, 2.7, 0.5, 0.3],
    "27 Dresses" : [2.5, 4.5, 4.0, 2.4, 0.8],
    "Suits" : [0.6, 3.5, 3.5, 1.5, 0.9],
    "Black Hat" : [0.8, 3.0, 4.4, 0.5, 0.8],
    "Matrix" : [3.6, 0.2, 0.7, 1.6, 0.5]
}

ans = {k : round(sum(v)/len(v),1) for k,v in r.items()}


print(f"{"Movie Name":<15}{"Average Rating"}")
for k,v in ans.items():
    print(f"{k : <15}{v}")
```

```python
# Student Report Card System  Use nested dictionaries to store student → subjects → marks.  Write a function to print each student’s average.

s = {
    "Darshan" : {"Maths" : 90, "Science" : 89, "English" : 60, "Hindi" : 78, "Computer" : 89},
    "Jay" : {"Maths" : 99, "Science" : 59, "English" : 30, "Hindi" : 38, "Computer" : 83},
    "Meet" : {"Maths" : 70, "Science" : 39, "English" : 50, "Hindi" : 78, "Computer" : 59},
    "Deep" : {"Maths" : 50, "Science" : 89, "English" : 50, "Hindi" : 38, "Computer" : 39},
    "Priyank" : {"Maths" : 80, "Science" : 39, "English" : 60, "Hindi" : 73, "Computer" : 83}
}


print(f" | {"Name" :<12} | {"Maths":<12} | {"Science":<12} | {"English":<12} | {"Hindi":<12} | {"Computer":<12} | {"Average Mark":<12} |")
for i in s:
    print(f" | {i :<12}", end="")
    for j in s[i]:
        print(f" | {s[i][j] :<12}", end="")
    print(f" | {sum(s[i].values()) / len(s[i].values()) :<12} | ")
```

```python
# Employee Salary Processor  Use filter() to select employees with salary above a threshold, and reduce() to find total company salary expense.

from functools import reduce

e = {
    "Darshan" : 15000,
    "Jay" : 12000,
    "Meet" : 40000,
    "Deep" : 80000,
    "Priyank" : 50000
}

threshold = 20000
ans = list([i for i in e if e[i] in filter(lambda x : x > threshold, e.values())])
print("Employee's whose salary is above threshold :- ",ans)

expense = reduce(lambda x, y : x + y, e.values())
print("\nTotal company salary expense :-",expense)
```

```python
# Crypto Price Tracker  Store crypto names and prices in a dictionary. Write a function that returns the name of the cryptocurrency with the highest price.

crypto = {
    "BitCoin" : 120,
    "Ethereum" : 89,
    "Tether" : 78,
    "Solana" : 20,
    "Cardano" : 120,
}
def highest(a):
    high = max(a.values())
    return [i for i in a if a[i] == high]

for i in crypto:
    print(i,crypto[i])

print("\nHighest Price :-",highest(crypto))
```
