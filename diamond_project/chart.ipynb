```python
plt.figure(figsize = [12,6])
plt.hist(data['price'], bins = 120, color = "purple")
plt.xlabel("Price in USD")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize = [12,6])
plt.plot(data['price'],data['carat'])
plt.xlabel("Price")
plt.ylabel("Carat")
plt.title("Price V/S Carat")
plt.show()
```

```python
plt.figure(figsize = [12,6])
plt.plot(data['price'],data['depth'])
plt.xlabel("Price")
plt.ylabel("Depth")
plt.title("Price V/S Carat")
plt.show()
```

```python
cutCounts = data['cut'].value_counts()

plt.figure(figsize = [12,6])
cutCounts.plot(kind = "line", color = "orange", ls = "-", marker = 'o', ms = "12", mec = "orange", mfc = "white")
plt.title("Distribution of Diamont cut")
plt.xlabel("Diamond cut")
plt.ylabel("Counts")
plt.xticks(rotation = 0)
plt.show()
```

```python
cutCounts = data['cut'].value_counts()

plt.figure(figsize = [12,6])
cutCounts.plot(kind = "bar", color = "orange")
plt.title("Distribution of Diamont cut")
plt.xlabel("Diamond cut")
plt.ylabel("Counts")
plt.xticks(rotation = 0)
plt.show()
```

```python
cutCounts = data['cut'].value_counts()
plt.figure(figsize = [12,6])
cutCounts.plot(kind = "pie", autopct = "%1.1f%%",colors = ["skyblue", "lightgreen", "orange", "lightcoral", "gold"])
plt.title("Distribution of Diamont cut")
plt.show()
```

```python
colorCounts = data['color'].value_counts()

plt.figure(figsize = [12,6])
colorCounts.plot(kind = "line", color = "gold", ls = "--", marker = "d", ms = 10, mfc = "white", mec = "gold")
plt.title("Distribution of Diamond color")
plt.xlabel("Diamond color")
plt.ylabel("Count")
plt.show()
```

```python
colorCounts = data['color'].value_counts()

plt.figure(figsize = [12,6])
colorCounts.plot(kind = "bar", color = "gold")
plt.title("Distribution of Diamond color")
plt.xlabel("Diamond color")
plt.ylabel("Count")
plt.xticks(rotation = 0)
plt.show()
```

```python
colorCounts = data['color'].value_counts()

plt.figure(figsize = [12,6])
colorCounts.plot(kind = "pie", autopct = "%1.1f%%", colors = ["skyblue", "gold", "lightgreen", "lightcoral", "orange", "hotpink", "forestgreen"])
plt.title("Distribution of Diamond color")
plt.show()
```

```python
clarityCounts = data['clarity'].value_counts()

plt.figure(figsize = [12,6])
clarityCounts.plot(kind = "line", marker = "s", ls = ":", mfc = "white", mec = "black", color = "black")
plt.title("Distribution of Diamond Clarity")
plt.xlabel("Clarity")
plt.ylabel("Count")
plt.show()
```

```python
clarityCounts = data['clarity'].value_counts()

plt.figure(figsize = [12,6])
clarityCounts.plot(kind = "bar", color = "black")
plt.title("Distribution of Diamond Clarity")
plt.xlabel("Clarity")
plt.ylabel("Count")
plt.show()
```

```python
clarityCounts = data['clarity'].value_counts()

plt.figure(figsize = [12,6])
clarityCounts.plot(kind = "pie", 
                   autopct = "%1.1f%%",
                   colors = ["skyblue", "gold", "lightgreen", "lightcoral", "orange", "hotpink", "forestgreen", "darkred"],
                   labeldistance = 1.1,  
                   pctdistance = 0.8)
plt.title("Distribution of Diamond Clarity")
plt.show()
```

```python
plt.figure(figsize = [12,6])
plt.scatter(data['x'], data['y'])
plt.title("X V/S Y", loc = "center") 
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
