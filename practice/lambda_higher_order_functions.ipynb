## Lambda, Higher-Order Functions

Create a lambda function that squares a number.

```python
sqaure_num = lambda x : x ** 2

num = int(input("Enter the number :-"))

print(f"Sqaure of {num} :- {sqaure_num(num)}")
```

Create a lambda function that returns the cube of a number.

```python
cube_num = lambda x : x ** 3

num = int(input("Enter the number :-"))

print(f"Cube of {num} :- {cube_num(num)}")
```

Create a lambda function that checks whether a number is even.

```python
is_even = lambda x : True if x % 2 == 0 else False

num = int(input("Enter the number :-"))

print(f"{num} is {"Even" if is_even(num) else "Odd"}")
```

Create a lambda function that returns the larger of two numbers.

```python
max_num = lambda x, y : max(x,y)

num1 = int(input("Enter the first number :-"))
num2 = int(input("Enter the second number :-"))

print(f"Larger number :- {max_num(num1, num2)}")
```

Create a lambda function that returns the length of a string.

Use map() to convert a list of integers into squares.

```python
l = [1, 2, 3, 4, 5]

square_list = list(map(lambda x : x ** 2, l))

print(f"List :- {l}")
print(f"Square List :- {square_list}")
```

Use map() to convert a list of strings into uppercase.

```python
name = ["darshan", "deep", "dipak", "jay", "meet"]

upper_name = list(map(lambda x : x.upper() ,name))

print(f"Name :- {name}")
print(f"Upper Case :- {upper_name}")
```

Use map() to find the length of every word.

```python
words = ["this", "is", "my", "pen", "book"]

length_words = list(map(lambda x : len(x), words))

print(f"| {"Word" :<5} | {"Length":<6} |")
for i,j in zip(words, length_words):
    print(f"| {i :<5} | {j :<6} |")
```

Convert Celsius temperatures into Fahrenheit using map().

```python
celsius = [10, 20, 56, 89.82, 65.14]

fahrenheit = list(map(lambda x : x * 9 / 5 + 32, celsius))

print(f"{"Celsius" :<10} {"Fahrenheit" :<10}")

for i, j in zip(celsius, fahrenheit):
    print(f"{i :<10} {j :<10}")
```

Sort dictionary items by value.

```python
students = {
    "Darshan" : 100,
    "Aryan" : 90,
    "Meet" : 78,
    "Priyank" : 70,
    "Jay" : 69
}

ans = dict(sorted(students.items(), key = lambda x : x[1]))

for i in students:
    print(f"{i :<10} :- {students[i]}")

print("\nSorted by value")
for i in ans:
    print(f"{i :<10} :- {ans[i]}")
```

Sort students by marks descending.

```python
students = {
    "Darshan" : 110,
    "Aryan" : 91,
    "Meet" : 78,
    "Priyank" : 80,
    "Jay" : 100
}

ans = dict(sorted(students.items(), key = lambda x : x[1], reverse = True))

for i in students:
    print(f"{i :<10} :- {students[i]}")

print("\nSorted by marks (high to low)")
for i in ans:
    print(f"{i :<10} :- {ans[i]}")
```

Find sum of all numbers using reduce.

```python
from functools import reduce

l = [6, 4, 9, 30, 3, 1, 75, 45, 8, 7, 5, 90, 15, 60, 2]

ans = reduce(lambda x,y : x + y, l)

print(f"Numbers :- {l}")
print(f"Sum :- {ans}")
```

Find product of all numbers.

```python
from functools import reduce

l = [1, 2, 3, 4, 5]

ans = reduce(lambda x,y : x * y, l)

print(f"Numbers :- {l}")
print(f"Product :- {ans}")
```

Find largest number using reduce.

```python
from functools import reduce

l = [6, 4, 9, 30, 3, 1, 75, 45, 8, 7, 5, 90, 15, 60, 2]

ans = reduce(lambda x,y : x if x > y else y, l)

print(f"Numbers :- {l}")
print(f"Largest :- {ans}")
```

Find smallest number using reduce.

```python
from functools import reduce

l = [6, 4, 9, 30, 3, -10, 75, 45, 8, 7, 5, 90, 15, 60, 2]

ans = reduce(lambda x,y : x if x < y else y, l)

print(f"Numbers :- {l}")
print(f"Smallest :- {ans}")
```

Concatenate a list of strings using reduce.

```python
words = ["this", "is", "my", "pen."]

ans = reduce(lambda x, y : x + " " + y, words)

print(f"Words :- {words}")
print(f"Ans :- {ans}")
```

Transpose a matrix.

```python
l = [[1,2,9], [3,4,7], [5,6,8]]

r , c = len(l), len(l[0])
ans = [[l[j][i] for j in range(c)] for i in range(r)]

for i in l:
    print(i)

print("\nTranspose :-")
for i in ans:
    print(i)
```
