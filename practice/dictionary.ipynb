```python
# Q1. Create a dictionary of 5 students with their marks. Write a program to display all keys and all values separately.
students = {
    "Darshan" : 100,
    "Aryan" : 90,
    "Meet" : 78,
    "Priyank" : 70,
    "Jay" : 69
}

print("KEY :-")
for i in students.keys():
    print(i)

print("\nVALUES :-")
for i in students.values():
    print(i)

for i in students.items():
    print(i[0],i[1])
```

```python
# Q7. Write a program to count how many values in a dictionary are even and how many are odd.

students = {
    "Darshan" : 100,
    "Aryan" : 90,
    "Meet" : 79,
    "Priyank" : 70,
    "Jay" : 69
}

ce = co = 0

for i in students:
    if students[i] % 2 == 0:
        ce += 1 
    else:
         co += 1

print("Even :-",ce)
print("Odd :-",co)
```

```python
# Q10. Write a program to count the frequency of each word in a given sentence.

text = input("Enter the name :-")
words = text.split()
ans = {}

for i in words:
    ans[i] = words.count(i)

print(text)
print(ans)
```

```python
# Q17. Write a function to add a new student record. Make sure duplicate IDs are not allowed.

def addData(students, id):
    name = input("Enter the name :-")
    mark = int(input("Enter the mark :-"))
    students[id] = {"name" : name, "marks" : mark}


students ={
    1: {"name": "Rahul", "marks": 80},
    2: {"name": "Priya", "marks": 65},
    3: {"name": "Amit", "marks": 90},
    4: {"name": "Aman", "marks": 55},
    5: {"name": "Darshan", "marks": 49}
}

id = int(input("Enter the id :-"))

if id not in students:
    addData(students, id)

    for i in students:
        print(i, students[i]["name"], students[i]["marks"])

else:
    print("Duplicate id not allowed.")
    
```

```python
# Q18. Write a function to update student details based on ID. The function should allow updating name, marks, or both.

def updateData(students, id):
    name = input("Enter the name :-")
    mark = int(input("Enter the mark :-"))
    students[id] = {"name" : name, "marks" : mark}
    

students ={
    1: {"name": "Rahul", "marks": 80},
    2: {"name": "Priya", "marks": 65},
    3: {"name": "Amit", "marks": 90},
    4: {"name": "Aman", "marks": 55},
    5: {"name": "Darshan", "marks": 49}
}

id = int(input("Enter the id :-"))

if id in students:
    updateData(students, id)

    for i in students:
        print(i, students[i]["name"], students[i]["marks"])

else:
    print("Id not found")
```

```python
# Q20. Write a program to find the student who has scored the highest marks.

students ={
    1: {"name": "Rahul", "marks": 80},
    2: {"name": "Priya", "marks": 65},
    3: {"name": "Amit", "marks": 90},
    4: {"name": "Aman", "marks": 90},
    5: {"name": "Darshan", "marks": 49}
}

hm = max([students[i]["marks"] for i in students])

for i in students:
    if students[i]["marks"] == hm:
        print(i, students[i]["name"], students[i]["marks"])
```

```python
# Q21. Write a program to assign grades to each student based on marks:
# Above 75 → Grade A
# Between 50 and 75 → Grade B
# Below 50 → Grade C

students ={
    1: {"name": "Rahul", "marks": 80},
    2: {"name": "Priya", "marks": 65},
    3: {"name": "Amit", "marks": 45},
    4: {"name": "Aman", "marks": 90},
    5: {"name": "Darshan", "marks": 49}
}

for i in students:
    grade = ""
    m = students[i]["marks"]
    
    if  m > 75:
        grade = "A"
    elif m > 50:
        grade = "B"
    else:
        grade = "C"
    
    students[i]["grade"] = grade
    
for i in students:
    print(i, students[i]["name"], students[i]["marks"], students[i]["grade"])
```
