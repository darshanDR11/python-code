#### Write a program that reads an integer and prints whether the number is odd or even.

```python
# odd even
n = int(input("Enter the number"))
if n % 2 == 0:
    print(f"{n} is even.")
else:
    print(f"{n} is odd.")
```

#### Write a program to find and print all factors of a given positive integer.

```python
n = int(input("Enter the number"))
factors = [i for i in range(1,n+1) if n % i == 0]
print(f"Factors of {n} : {factors}")
```

#### Write a program to compute the factorial of a given integer using a loop.

```python
# factorial
n = int(input("Enter the number"))
ans = 1

for i in range(1, n + 1):
    ans = ans * i

print("Factorial :- ",ans)
```

#### Write a program to check whether a given number is prime or not.

```python
n = int(input("Enter the number"))

for i in range(2,n):
    if n % i == 0:
        print(n,"is not a prime number.")
        break
else:
    print(n,"is a prime number.")
```

#### Write a program that prints the Fibonacci series up to N terms (where N is given by the user).

```python
n = int(input("Enter the total number you want :-"))
a, b = -1 , 1

print("Fibonaci series :- ", end = "")
for i in range(n):
    print(a+b, end = " ")
    a,b = b, a+b

```

#### Write a program to check whether a given number is a perfect number or not (sum of proper divisors equals the number).

```python
n = int(input("Enter the number :-"))
sum = 0

for i in range(1,n):
    if n % i == 0:
        sum += i

if sum == n:
    print(f"{n} is perfect number.")
else:
    print(f"{n} is not perfect number.")
```

#### Write a program to reverse the digits of a given number and display the reversed number.

```python
num = int(input("Enter the number :-"))
reversed = 0
num1 = num

while num1 > 0:
    rem = num1 % 10
    reversed = reversed * 10 + rem
    num1 = num1 // 10

print("Original number :-",num)
print("Reversed number :-",reversed)
```

#### Write a program to check whether a given number is a palindrome or not. 

```python
n = int(input("Enter the number"))

if str(n) == str(n)[::-1]:
    print(f"{n} is palindrome number.")
else:
    print(f"{n} is not palindrome number.")
```

#### Write a program to check whether a given number is an Armstrong number (sum of its digits each raised to the power of number of digits).

```python
num = int(input("Enter the number"))

power = len(str(num))

num1 = num
sum = 0

while num1 > 0:
    rem = num1 % 10
    sum = sum + (rem ** power)
    num1 = num1 // 10

if sum == num:
    print(f"{num} is armstrong number.")
else:
    print(f"{num} is not armstrong number.")
```

#### Write a program to compute the sum of digits of a number and also count how many digits it has.

```python
num = int(input("Enter the number"))

num1 = num
sum = 0
count = 0

while num1 > 0:
    rem = num1 % 10
    sum = sum + rem
    num1 = num1 // 10
    count = count + 1

print(f"Sum of digits :- {sum}")
print(f"Total digits :- {count}")
```

#### Check whether a number is a Harshad/Niven number (divisible by sum of digits).

```python
num = int(input("Enter the number :-"))

num1 = num
sum1 = 0
while num1 > 0:
    rem = num1 % 10
    sum1  = sum1 + rem
    num1 = num1 // 10

if num % sum1 == 0:
    print(f"{num} is Harshad number.")
else:
    print(f"{num} is not Harshad number.")
```

#### Check whether a number is a Strong number (sum of factorials of digits equals the number).

```python
import math
num = int(input("Enter the number :-"))

num1 = num
sum1 = 0

while num1 > 0:
    rem = num1 % 10
    sum1 = sum1 + math.factorial(rem)
    num1 = num1 // 10

if sum1 == num:
    print(f"{num} is strong number.")
else:
    print(f"{num} is not strong number.")
```

#### Check whether a number is a Spy number (sum of digits equals product of digits).

```python
from functools import reduce
n = int(input("Enter the number :-"))
d = [int(i) for i in str(n)]

sum1 = reduce(lambda x, y : x + y, d)
product1 = reduce(lambda x, y : x * y, d)

if sum1 == product1:
    print(f"{n} is spy number.")
else:
    print(f"{n} is not spy number.")
```

#### Check whether a number is a Disarium number (digit^position sums to the number). (Advanced)

```python
n = int(input("Enter the number :-"))

digits = [int(i) for i in str(n)]

sum1 = sum([digits[i] ** (i+1) for i in range(len(digits))])

if sum1 == n:
    print(f"{n} is Disarium number.")
else:
    print(f"{n} is not Disarium number.")
```

#### Given N, print the sum of even digits and the sum of odd digits.

```python
n = int(input("Enter the number :-"))

n1 = n
oddSum , evenSum = 0, 0

while n1 > 0:
    rem = n1 % 10
    if rem % 2 == 0:
        evenSum += rem
    else:
        oddSum += rem
    n1 = n1 // 10

print(f"Sum of odd digits :- {oddSum}")
print(f"Sum of even digits :- {evenSum}")
```
