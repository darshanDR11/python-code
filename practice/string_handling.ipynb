#### Reverse a string (using slicing).

```python
name = input("Enter the name :-")
reversed = name[::-1]
print(reversed)
```

#### Reverse a string (without using slicing).

```python
name = input("Enter the name :-")
reversed = ""
for i in range(len(name)-1, -1, -1):
    reversed += name[i]

print(reversed)
```

#### Count vowels and consonants (constants) in a string.

```python
name = input("Enter the name :-")

cv = cc = 0

for i in name:
    if i in "AEIOUaeiou":
        cv += 1
    else:
        cc += 1


print(f"Total vowels :- {cv}")
print(f"Total constants :- {cc}")
```

#### Count alphabets, digits, and special characters in a text.

```python
text = input("Enter the text :-")

cd = csc = ca = 0

for i in text:
    if i.isalpha():
        ca += 1
    elif i.isdigit():
        cd += 1
    else :
        csc += 1

print("Total Alphabets :-",ca)
print("Total Digits :-",cd)
print("Total Special Characters :-",csc)
```

#### Swap case of letters in a string using built-in methods.

```python
name = input("Enter the name :-")

ans = []
for i in name:
    if i.isalpha():
        if i.isupper():
            ans.append(i.lower())
        else:
            ans.append(i.upper())
    else:
        ans.append(i)

print("Original Name :-",name)
print("After case swapping :-","".join(ans))
```

#### Swap case of letters in a string without using built-in case functions (use ASCII/ord).

```python
name = input("Enter the name :-")

ans = []
for i in name:
    if ord(i) >= 65 and ord(i) <= 90:
        ans.append(chr(ord(i) + 32))
    elif ord(i) >= 97 and ord(i) <= 122:
        ans.append(chr(ord(i) - 32))
    else:
        ans.append(i)

print("Original Name :-",name)
print("After case swapping :-","".join(ans))
```

#### Compress a string (e.g., consecutive same characters → character followed by count).

```python
text = input("Enter the text :-")

ans = []
l = len(text)

i = 0
while i < l:
    j = i
    c = 0
    while j < l and text[j] == text[i]:
        c += 1
        j += 1
    ans.append(f"{text[i]}{c}")
    i = i + c 

print("Compressed text :-","".join(ans))
```

#### From a sentence, find the longest words and shortest words.

```python
text = input("Enter the text :-")

words = text.split(" ")

max_len = max([len(word) for word in words])
min_len = min([len(word) for word in words])

longest_words = []
shortest_words = []

for i in words:
    if len(i) == max_len:
        longest_words.append(i)
    if len(i) == min_len:
        shortest_words.append(i)

print("Longest Words :-",longest_words)
print("Shortest Words :-",shortest_words)
```

#### Find the maximum occurring character in a string (if tie print all characters).

```python
name = input("Enter the name :-")

ans = {i : name.count(i) for i in name if i != " "}

max_occur = max(ans.values())
c = []

for i in ans:
    if ans[i] == max_occur:
        c.append(i)

if len(c) > 1:
    print("All characters :-", list(ans.keys()))
else:
    print("Maximum occuring character :-",c)
```

#### Given a string, determine if it is “pangram” (contains all letters a-z).

```python
s1 = input("Enter the text :-")

isPangram = True

for i in range(65, 91):
    if chr(i) not in s1 and chr(i).lower() not in s1:
        isPangram = False

if isPangram:
    print(f"{s1} is pangram.")
else:
    print(f"{s1} is not pangram.")
```

#### Check whether two strings are anagrams.

```python
s1 = input("Enter the string 1 :-").replace(" ","")
s2 = input("Enter the string 2 :-").replace(" ","")

is_anagram = True

for i in s1:
    if i not in s2:
        is_anagram = False
        break

if is_anagram:
    print(f"{s1} , {s2} are anagrams string.")
else:
    print(f"{s1}, {s2} are not anagrams string.")
```

#### Implement your own version of split() without using split().

```python
def split_string(s):
    ans = []
    word = ""
    
    for i in s:
        if i != " ":
            word = word + i
        else:
            ans.append(word)
            word = ""
    ans.append(word)
    return ans    

print(split_string("My name is Deep"))
print(split_string("1 2 3 4  5 6"))
print(split_string("this is my pen"))
```

#### Implement your own version of join() without using join().

###### 1 approach using reduce and lambda function.

```python
from functools import reduce
def join_value(value, delimeter):
    ans = reduce(lambda x,y : x + delimeter +  y , value)
    return ans 
        
print(join_value(['My', 'name', 'is', 'Deep'], " "))
print(join_value(['1', '2', '3', '4', '5', '6'], ","))
```

###### 2 approach using for loop.

```python
def join_value(value, delimeter):
    ans = value[0]
    for i in range(1, len(value)):
        ans = ans + delimeter + value[i]
    return ans
        
print(join_value(['My', 'name', 'is', 'Deep'], " "))
print(join_value(['1', '2', '3', '4', '5', '6'], ","))
```

#### Implement your own version of count().

```python
def count_c(s, v):
    c = 0
    l = len(s)

    for i in range(l):
        for j in range(i+1,l+1):
            if s[i:j] == v:
                c += 1
    return c
    
print(count_c("Darsharn","ar"))
print(count_c("Deep", "e"))
```

#### Find the longest substring containing only unique characters.

```python
name = input("Enter the name :-")

l = len(name)

# substring = ["".join(i for i in name[i:j]) for i in range(l) for j in range(i+1,l+1)]
substring = [name[i:j] for i in range(l) for j in range(i+1,l+1) if len(name[i:j]) == len(set(name[i:j]))]


print("Substring with unique characters :-")
for i in substring:
    print(i)

print("\nLongest substring with unique charaters :- ",sorted(substring, key = len, reverse = True)[0])    
```

#### Find the longest substring containing all vowels.

##### list comprehension

```python
s = input("Enter the string :-")
l = len(s)
substring = [s[i:j] for i in range(l) for j in range(i+1,l+1) if all(ch in "AEIOUaeiou" for ch in s[i:j])]
print("Longest substring with all vowels :-",sorted(substring, key = len, reverse = True)[0])
```

##### using for loop

```python
s = input("Enter the string :-")
l = len(s)
substring = []

for i in range(l):
    for j in range(i+1, l+1):
        a = s[i:j]
        if all(ch in "AEIOUaeiou" for ch in a):
            substring.append(a)
                
print("Shortest string with all vowels :-",sorted(substring, key = len, reverse = True)[0])
```
