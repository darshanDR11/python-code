```python
# 4. Write a function matrix_max(matrix) that finds and returns the maximum element present anywhere in a given 2D matrix. For example, if the input is [[3, 99, 7], [12, 45, 2], [88, 5, 6]], the output should be 99. If the matrix is empty, return None.

def matrix_max(matrix):
    if matrix:
        matrix = [j for i in matrix for j in i]
        return max(matrix) 
    else:
        return None
matrix = [[3, 99, 7], [12, 45, 2], [88, 5, 6]]

print("Maximum number :-",matrix_max(matrix))
```

```python
# 5. Write a function matrix_min(matrix) that finds and returns the minimum element present anywhere in a given 2D matrix. For example, if the input is [[8, 3, 9], [1, 4, 7], [2, 0, 6]], the output should be 0. Handle the case of an empty matrix gracefully.

def matrix_min(matrix):
    if matrix:
        matrix = [j for i in matrix for j in i]
        return min(matrix) 
    else:
        return None
matrix = [[8, 3, 9], [1, 4, 7], [2, 0, 6]]

print("Minimum number :-",matrix_min(matrix))
```

```python
# 6. Write a function row_wise_max(matrix) that returns a list containing the maximum element from each row of a 2D matrix. For example, if matrix = [[1,5,3], [9,2,8], [4,6,7]], the output should be [5, 9, 7]. If any row is empty, treat its maximum as None.


def row_wise_max(matrix):
    maximum = []
    for i in matrix:
        maximum.append(max(i)) 
    return maximum
        

matrix = [[8, 3, 9], [2,3,4], [2, 0, 6]]

print("Maximum number :-",row_wise_max(matrix))
```

```python
# 9. Write a function all_frequencies(lst) that returns a dictionary where keys are unique numbers from the list and values are their frequencies. For example, if lst = [4, 2, 4, 3, 2, 2], the output should be {4: 2, 2: 3, 3: 1}. Also print the dictionary in a readable format.

def all_frequencies(lst):
    ans = {}

    unique = set([i for i in lst])
    
    for i in unique:
        ans[i] = lst.count(i)

    return ans

lst = [4, 2, 4, 3, 2, 2]
ans = all_frequencies(lst)

print("Digit \t Count")
for i in ans:
    print(i," \t ",ans[i])
```

```python
# 10. Write a function remove_duplicates_preserve_order(lst) that returns a new list containing only the first occurrence of each element, preserving the original order. For example, [4, 2, 4, 3, 2, 5, 4] should become [4, 2, 3, 5]. Do not use set() directly.

def remove_duplicates(lst):
    ans = []
    for i in lst:
        if i not in ans:
            ans.append(i)
    return ans

lst = [4, 2, 4, 3, 2, 5, 4] 
print(remove_duplicates(lst))
```

```python
# 11. Write a function replace_first_last_occurrence(lst, target, replacement) that replaces only the first and last occurrence of the target value in the list. For example, if lst = [1, 2, 3, 2, 4, 2, 5], target = 2, replacement = 99, the output should be [1, 99, 3, 2, 4, 99, 5]. If the target appears only once, replace that single occurrence.

def replace_first_last_occurrence(l,t,r):
    first = last = -1

    for i in range(len(l)):
        if l[i] == t:
            if first == -1:
                first = i
            else:
                last = i

    l[first] = l[last] = r
    
    return l

l = [1, 2, 3, 2, 4, 2, 5]
t = 2
r = 99

ans = replace_first_last_occurrence(l,t,r)

print(ans)
```

```python
# 12. Write a function replace_middle_occurrences(lst, target, replacement) that replaces all occurrences of the target value except the first and last occurrence. For example, [1, 2, 3, 2, 4, 2, 5, 2] with target 2 and replacement 99 should become [1, 2, 3, 99, 4, 99, 5, 2]. If less than 3 occurrences, make no changes.

def replace_middle_occurrences(l,t,r):
    first = last = -1

    for i in range(len(l)):
        if l[i] == t:
            if first == -1:
                first = i
            else:
                last = i

    for i in range(len(l)):
        if i>first and i<last and l[i] == t:
            l[i] = r    
    return l

l = [1, 2, 3, 2, 4, 2, 5, 2]
t = 2
r = 99

ans = replace_middle_occurrences(l,t,r)

print(ans)
```

```python
# 13. Write a function replace_all_occurrences(lst, target, replacement) that replaces every occurrence of the target value with the replacement value. For example, [1, 2, 3, 2, 4, 2] with target 2 and replacement 99 becomes [1, 99, 3, 99, 4, 99]. Return the modified list.

def replace_all_occurrences(l,t,r):
    l = [r if i == t else i for i in l]
    return l

l = [1, 2, 3, 2, 4, 2]
t = 2
r = 99

ans = replace_all_occurrences(l,t,r)

print(ans)
```

```python
# 18.  Write a function add_tuples(t1, t2) that takes two tuples of same length and returns a new tuple where each element is the sum of corresponding elements. Example: (1,2,3) + (4,5,6) = (5,7,9). Take both tuples as input from the user.

def add_tuples(t1, t2):
    return tuple([t1[i] + t2[i] for i in range(len(t1))])

a = input("Enter the value of tuple 1 :-")
b = input("Enter the value of tuple 2 :-")

t1 = tuple([int(i) for i in a.split()]) 
print(t1)

t2 = tuple([int(i) for i in b.split()]) 
print(t2)

sum = add_tuples(t1,t2)
print(sum)
```

```python
# 20.  Write a function find_common(t1, t2) that takes two tuples and prints the elements that appear in both tuples (without duplicates). Example: (1,2,3,4) and (3,4,5,6) → Common elements: 3, 4.
 
def find_common(t1,t2):
    ans = []
    for i in t1:
        if i in t2:
            ans.append(i)
    ans = tuple(ans)
    print(ans)
    

a = input("Enter the value of tuple 1 :-")
b = input("Enter the value of tuple 2 :-")

t1 = tuple([int(i) for i in a.split()]) 
print(t1)

t2 = tuple([int(i) for i in b.split()]) 
print(t2)

find_common(t1,t2)
```

```python
# 22.  Write a function second_largest(tup) that finds and returns the second largest element from a tuple without using sorting. Example: (10, 5, 8, 20, 15) → 15. Assume all elements are unique.

def second_largest(t1):
    l = max(list(t1))
    sl = -1
    for i in t1:
        if i < l and i > sl:
            sl = i

    return sl

a = input("Enter the value of tuple 1 :-")
t1 = tuple([int(i) for i in a.split()]) 
print(t1)

sl = second_largest(t1)
print(sl)  
```

```python
# 23. Write a user-defined function set_union(set1, set2) that returns the union of two sets without using the | operator or union() method. For example, if set1 = {1,2,3} and set2 = {3,4,5}, the output should be {1,2,3,4,5}. Use only loops and conditions.
 
def set_union(s1, s2):
    ans = set(list(s1) + list(s2))
    return ans

a = input("Enter the value of set1 :- ")
b = input("Enter the value of set2 :- ")

s1 = set([int(i) for i in a.split()])
s2 = set([int(i) for i in b.split()])

print(s1,s2)

ans = set_union(s1,s2)
print("Ans :-",ans)
 
```

```python
# 24. Write a user-defined function set_intersection(set1, set2) that returns the intersection of two sets without using the & operator or intersection() method. For example, {1,2,3,4} and {3,4,5,6} should return {3,4}.

def set_intersection(s1, s2):
    ans = {i for i in s1 if i in s2}
    return ans if ans else None

a = input("Enter the value of set1 :- ")
b = input("Enter the value of set2 :- ")

s1 = set([int(i) for i in a.split()])
s2 = set([int(i) for i in b.split()])

print(s1,s2)

ans = set_intersection(s1,s2)
print("Ans :-",ans)
```

```python
# 25. Write a user-defined function set_difference(setA, setB) that returns the set difference setA - setB (elements present in setA but not in setB) without using the - operator or difference() method. For example, {1,2,3,4} - {3,4,5} should return {1,2}.

def set_difference(s1, s2):
    ans = {i for i in s1 if i not in s2}
    return ans if ans else None

a = input("Enter the value of set1 :- ")
b = input("Enter the value of set2 :- ")

s1 = set([int(i) for i in a.split()])
s2 = set([int(i) for i in b.split()])

print(s1,s2)

ans = set_difference(s1,s2)
print("Ans :-",ans)
```

```python
# 26. Write a user-defined function symmetric_difference(set1, set2) that returns elements present in either set but not in both, without using ^ operator or symmetric_difference() method. For example, {1,2,3,4} and {3,4,5,6} should return {1,2,5,6}.

def sd(s1, s2):
    ans = set(list({i for i in s1 if i not in s2}) + list({i for i in s2 if i not in s1}))
    return ans if ans else None

a = input("Enter the value of set1 :- ")
b = input("Enter the value of set2 :- ")

s1 = set([int(i) for i in a.split()])
s2 = set([int(i) for i in b.split()])

print(s1,s2)

ans = sd(s1,s2)
print("Ans :-",ans)
```

```python
# 27. Write a user-defined function is_subset(setA, setB) that returns True if setA is a subset of setB, otherwise False. Do not use the <= operator or issubset() method. For example, {1,2} is a subset of {1,2,3} → True.

def is_subset(s1,s2):
    for i in s2:
        if i not in s1:
            return False
    return True


a = input("Enter the value of set1 :- ")
b = input("Enter the value of set2 :- ")

s1 = set([int(i) for i in a.split()])
s2 = set([int(i) for i in b.split()])

print(s1,s2)

ans = is_subset(s1,s2)
print("Ans :-",ans)
```
