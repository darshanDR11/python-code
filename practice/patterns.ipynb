```python
r = int(input("Enter the row-size :-"))

for i in range(1, r+1):
    for j in range(1, r+1):
        print("* " if i == 1 or i == r or j == 1 or j == r else "  ", end = "")
    print(" ")
```

```python
r = int(input("Enter the row-size :-"))

for i in range(1, r+1):
    for j in range(1, i+1):
        print("* " if j == 1 or j == i or i == r else "  ", end = "")
    print(" ")
```

```python
r = int(input("Enter the row-size :-"))

for i in range(1, r+1):
    for j in range(1, r+2-i):
        if j == 1 or i == 1 or j == r + 1 - i:
            print("* ", end = "")
        else:
            print("  ", end = "")
    print(" ")
```

```python
r = int(input("Enter the row-size :-"))

for i in range(1, r+1):
    print("  " * (r - i), end = "")
    for j in range(1, i * 2):
        if j == 1 or j == i * 2 - 1 or i == r:
            print("* ", end = "")
        else:
            print("  ", end = "")
    print(" ")
```

```python
r = int(input("Enter the row-size :-"))

for i in range(1, r+1):
    for j in range(1, r * 2):
        if j < i:
            print("  ", end = "")
        elif j == i or i == 1 or j == r * 2 - i:
            print("* ", end = "")
        else:
            print("  ", end = "")
    print(" ")
```

```python
r = int(input("Enter the row-size :-"))
for i in range(1, r+1):
    a = 2
    print("  " * (r - i),end = "")
    for j in range(1, 2 * i):
        if j <= i:
            print(j, end = " ")
        else:
            print(j - a,end = " ")
            a = a + 2
    print()
```
